```python
# >>> FUNCTION THAT ACTUALLY PERFORMS STRATIFIED SAMPLING <<< #

def stratified_sampling(dsn, col, n):
    dsn_g = dsn.groupby(col)
    frame = pd.DataFrame()
    
    frame['size'] = dsn_g.size()
    frame['prop'] = dsn_g.size()/len(dsn)
    frame['samsize'] = np.round(frame.prop*n)
    frame['levels'] = frame.index
    frame.reset_index(inplace=True)
    
    dsn2 = pd.merge(dsn, frame, on=col, how='inner')
    group = dsn2.groupby(col)
    grouped = list(group)
    
    ds_main = pd.DataFrame()
    ds_main = pd.concat([ds_main])
    
    for i in range(len(grouped)):
        ds = grouped[i][1]
        val = int(pd.unique(ds['samsize'])[0])    
        dss = ds.sample(val, random_state=0, axis=0)
        ds_main = pd.concat([ds_main, dss])
    
    return ds_main

#######################################################
# *************************************************** #
#######################################################

# >>> MAIN INTERACTIVE FUNCTION THAT DEPLOYES THE PREVIOUS FUNCTION TO IMPLEMENT STRATIFIED SAMPLING <<< #

def implement_stratified(data):
    numones = len(data[data['resp']==1])
    numzeros = len(data[data['resp']==0])
    rate = numones/(numones+numzeros)
    
    print("******")
    print()
    print("Current event rate: ", np.round(rate, 2))
    print()
    print("******")
    print()
    nsize = int(input("What overall sample size do you want: "))
    print()
    print("******")
    print()
    
    rr = float(input("What event rate do you want in the sample: "))
    sam_ones = int(np.round(rr*nsize))
    sam_zeros = int(nsize - sam_ones)
    
    print()
    print("******")
    print()
    print("Number of ones in your sample: ", sam_ones)
    print("Number of zeros in your sample: ", sam_zeros)
    print()
    print("******")
    print()
    
    strata = input("What column to use for stratification: ")
    
    print()
    print("******")
    
    one_sampled = stratified_sampling(data, strata, sam_ones)
    zero_sampled = stratified_sampling(data, strata, sam_zeros)
    
    final_frame = pd.concat([one_sampled, zero_sampled])
    final_frame = final_frame.drop(['size', 'prop', 'samsize', 'levels'], axis=1).reset_index(drop=True)
    
    return final_frame
```
