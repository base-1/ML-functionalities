```python
# >>> FUNCTION THAT CALCULATES BASIC META DATA AND SUMMARY INFORMATION ABOUT YOUR DATASET <<< #
# IT OUTPUTS:
# 1. THE COLUMNS PRESENT IN THE DATA
# 2. THE DATA TYPES ASSOCIATED WITH THE COLUMNS 
# 3. NUMBER OF UNIQUE VALUES IN EACH COLUMN 
# 4. SUMMARY STATISTICS OF NUMERIC VARIABLES 
# 5. PERCENTAGE OF MISSING VALUES IN EACH COLUMN

def meta_data(dsn):
    var_names = []
    d_type = []
    num_levels = []
    var_min = []
    var_25 = []
    var_med = []
    var_mean = []
    var_75 = []
    var_max = []
    var_sd = []
    perc_miss = []
    
    cols = list(dsn.columns)
    var_names = cols
    d_type = list(dsn.dtypes)
    
    for i in range(len(cols)):
        num_levels.append(len(pd.unique(dsn[cols[i]])))
        perc_miss.append(np.round(dsn[cols[i]].isnull().sum()/len(dsn), 4))
        
        if (dsn.dtypes[cols[i]] in ['float64', 'int64']):
            var_min.append(np.round(dsn[cols[i]].min()))
            var_25.append(np.round(dsn[cols[i]].quantile(0.25, interpolation='nearest'), 2))
            var_med.append(np.round(dsn[cols[i]].quantile(0.50, interpolation='nearest'), 2))
            var_mean.append(np.round(dsn[cols[i]].mean(), 2))
            var_75.append(np.round(dsn[cols[i]].quantile(0.75, interpolation='nearest'), 2))
            var_max.append(np.round(dsn[cols[i]].max(), 2))
            var_sd.append(np.round(dsn[cols[i]].std(), 2))
        else:
            var_min.append(np.nan)
            var_25.append(np.nan)
            var_med.append(np.nan)
            var_mean.append(np.nan)
            var_75.append(np.nan)
            var_max.append(np.nan)
            var_sd.append(np.nan)
        
    meta = pd.DataFrame()
    meta['variable'] = var_names
    meta['data_type'] = d_type
    meta['num_levels'] = num_levels
    meta['minimum'] = var_min
    meta['percentile_25'] = var_25
    meta['median'] = var_med
    meta['mean'] = var_mean
    meta['percentile_75'] = var_75
    meta['max'] = var_max
    meta['sd'] = var_sd
    meta['missing_perc'] = perc_miss
    
    return meta
```
